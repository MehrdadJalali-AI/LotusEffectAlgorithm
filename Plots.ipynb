```python
import matplotlib.pyplot as plt
import numpy as np

# Defining the Hill-Valley function visually as in Figure 2
x = np.linspace(0, 10, 400)
y1 = np.sin(x) + np.cos(1.5 * x) + 3
y2 = np.sin(x) + np.cos(1.5 * x) + 2

# Two arbitrary points in search space
iq_x, iq_y = 2, y1[80]
ip_x, ip_y = 8, y1[320]

fig, ax = plt.subplots(figsize=(10, 6))

# Plot the two curves representing hills and valleys
ax.plot(x, y1, label='Fitness Landscape 1', color='blue')
ax.plot(x, y2, label='Fitness Landscape 2', color='red', linestyle='--')

# Highlight the points iq and ip
ax.scatter([iq_x, ip_x], [iq_y, ip_y], color='green', zorder=5)
ax.text(iq_x, iq_y + 0.2, 'iq', fontsize=12, ha='center', color='green')
ax.text(ip_x, ip_y + 0.2, 'ip', fontsize=12, ha='center', color='green')

# Drawing vertical lines to show the inner points as per hill-valley method
for xi in np.linspace(iq_x, ip_x, 5)[1:-1]:
    ax.axvline(xi, color='purple', linestyle=':', ymin=0.1, ymax=0.9)

# Annotations
ax.annotate('Internal Points Evaluated', xy=(5, 2), xytext=(5, 1),
            arrowprops=dict(facecolor='purple', arrowstyle='->'),
            fontsize=12, color='purple', ha='center')

# Setting labels and title similar to Figure 2
ax.set_xlabel('Search Space', fontsize=14)
ax.set_ylabel('Fitness', fontsize=14)
ax.set_title('Hill-Valley Method Visualization', fontsize=16)
ax.legend(fontsize=12)
ax.grid(True)

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Define fitness landscape with clearer oscillations
x = np.linspace(0.3, 1, 500)
y = 2 + np.sin(10 * x) * np.cos(5 * x)

fig, ax = plt.subplots(figsize=(10, 6))

# Plot the fitness landscape
ax.plot(x, y, color='black')

# Define refined points
points = {
    '$i_p$': (0.35, 2 + np.sin(3.5) * np.cos(1.75)),
    '$i_q$': (0.9, 2 + np.sin(9) * np.cos(4.5)),
    'interior1': (0.41, 2 + np.sin(4.1) * np.cos(2.05)),
    'interior2': (0.38, 2 + np.sin(3.8) * np.cos(1.9)),
    'interior3': (0.45, 2 + np.sin(4.5) * np.cos(2.25)),
    'interior4': (0.7, 2 + np.sin(7.0) * np.cos(3.5)),
    'interior5': (0.8, 2 + np.sin(8.0) * np.cos(4.0)),
    'interior6': (0.6, 2 + np.sin(6.0) * np.cos(3.0)),
    '$i_{p}~midpoint$': (0.53, 2 + np.sin(5.3) * np.cos(2.65)),
    '$i_{q}~midpoint$': (0.76, 2 + np.sin(7.6) * np.cos(3.8)),
}

# Plot points clearly
for label, (px, py) in points.items():
    ax.scatter(px, py, color='black', s=60, zorder=5)
    offset_y = 0.04 if 'midpoint' in label else 0.05
    ax.text(px, py + offset_y, label, fontsize=10, ha='center', va='bottom')

# Connect i_p and i_q with line
ip_x, ip_y = points['$i_p$']
iq_x, iq_y = points['$i_q$']
ax.plot([ip_x, iq_x], [ip_y, iq_y], linestyle='--', color='black', linewidth=1.9)

# Labels and title with improved clarity
ax.set_xlabel('$x$', fontsize=16)
ax.set_ylabel('$eval(x)$', fontsize=16)
#ax.set_title('Visualization of Hill-Valley Method', fontsize=18)

# Removing ticks for cleaner style
ax.set_xticks([])
ax.set_yticks([])

# Adding grid with subtler style
ax.grid(True, linestyle='--', alpha=0.4)

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define x-axis range
x = np.linspace(-2, 4, 200)

# Synthetic multimodal function (two peaks with a valley, more noise to match the image)
y = 0.5 * np.exp(-((x + 1) ** 2) / 0.5) + 1.2 * np.exp(-((x - 2) ** 2) / 0.5) + np.sin(2 * x)
y += np.random.normal(0, 0.1, size=x.size)  # Increased noise for jagged appearance

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(x, y, label='Fitness Function', color='black', linewidth=1)

# Define points i_p and i_q
i_p_x, i_q_x = -1.5, 2.5  # Positions for better spread
i_p_y = y[np.argmin(np.abs(x - i_p_x))]
i_q_y = y[np.argmin(np.abs(x - i_q_x))]

# Plot i_p and i_q with matching colors
plt.plot(i_p_x, i_p_y, 'o', label='$i_p$', color='blue', markersize=10)
plt.annotate('$i_p$', (i_p_x, i_p_y), textcoords="offset points", xytext=(0, -20), ha='center', fontsize=14)

plt.plot(i_q_x, i_q_y, 'o', label='$i_q$', color='red', markersize=10)
plt.annotate('$i_q$', (i_q_x, i_q_y), textcoords="offset points", xytext=(0, 20), ha='center', fontsize=14)

# Interior points between i_p and i_q (adjusted samples to reduce overlap)
samples = [0.15, 0.45, 0.85]  # Adjusted for better spacing
interior_x = [i_p_x + (i_q_x - i_p_x) * s for s in samples]
interior_y = [y[np.argmin(np.abs(x - ix))] for ix in interior_x]
colors = ['green', 'purple', 'orange']  # Colors matching the image
for ix, iy, color, label in zip(interior_x, interior_y, colors, ['interior1', 'interior3', 'interior5']):
    plt.plot(ix, iy, 'o', color=color, markersize=10)
    # Adjust label positions for clarity
    if label == 'interior1':
        plt.annotate(label, (ix, iy), textcoords="offset points", xytext=(-20, -20), ha='center', fontsize=14)
    elif label == 'interior3':
        plt.annotate(label, (ix, iy), textcoords="offset points", xytext=(-40, 20), ha='center', fontsize=14)
    else:  # interior5
        plt.annotate(label, (ix, iy), textcoords="offset points", xytext=(40, 20), ha='center', fontsize=14)

# Midpoint for i_p (shifted to avoid overlap with interior3)
mid_x = (i_p_x + i_q_x) / 2 + 0.3  # Shifted right to avoid interior3
mid_y = y[np.argmin(np.abs(x - mid_x))]
plt.plot(mid_x, mid_y, 'o', label='$i_p$ midpoint', color='cyan', markersize=10)
plt.annotate('$i_p$ midpoint', (mid_x, mid_y), textcoords="offset points", xytext=(40, 20), ha='center', fontsize=14)

# Midpoint for i_q (shifted to avoid overlap with i_q and interior5)
i_q_mid_x = i_q_x - 0.5  # Shifted further left
i_q_mid_y = y[np.argmin(np.abs(x - i_q_mid_x))]
plt.plot(i_q_mid_x, i_q_mid_y, 'o', label='$i_q$ midpoint', color='yellow', markersize=10)
plt.annotate('$i_q$ midpoint', (i_q_mid_x, i_q_mid_y), textcoords="offset points", xytext=(0, -20), ha='center', fontsize=14)

# Dashed line between i_p and i_q
plt.plot([i_p_x, i_q_x], [i_p_y, i_q_y], '--', color='black', linewidth=1)

# Labels and styling
#plt.title('Figure 2 - Model of the hill-valley function', fontsize=14)
plt.xlabel('x', fontsize=16)
plt.ylabel('eval(x)', fontsize=16)
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend(loc='lower right')
plt.tight_layout()

# Show the plot
plt.show()
```
