```python
import matplotlib.pyplot as plt
import numpy as np

# Defining the Hill-Valley function visually as in Figure 2
x = np.linspace(0, 10, 400)
y1 = np.sin(x) + np.cos(1.5 * x) + 3
y2 = np.sin(x) + np.cos(1.5 * x) + 2

# Two arbitrary points in search space
iq_x, iq_y = 2, y1[80]
ip_x, ip_y = 8, y1[320]

fig, ax = plt.subplots(figsize=(10, 6))

# Plot the two curves representing hills and valleys
ax.plot(x, y1, label='Fitness Landscape 1', color='blue')
ax.plot(x, y2, label='Fitness Landscape 2', color='red', linestyle='--')

# Highlight the points iq and ip
ax.scatter([iq_x, ip_x], [iq_y, ip_y], color='green', zorder=5)
ax.text(iq_x, iq_y + 0.2, 'iq', fontsize=12, ha='center', color='green')
ax.text(ip_x, ip_y + 0.2, 'ip', fontsize=12, ha='center', color='green')

# Drawing vertical lines to show the inner points as per hill-valley method
for xi in np.linspace(iq_x, ip_x, 5)[1:-1]:
    ax.axvline(xi, color='purple', linestyle=':', ymin=0.1, ymax=0.9)

# Annotations
ax.annotate('Internal Points Evaluated', xy=(5, 2), xytext=(5, 1),
            arrowprops=dict(facecolor='purple', arrowstyle='->'),
            fontsize=12, color='purple', ha='center')

# Setting labels and title similar to Figure 2
ax.set_xlabel('Search Space', fontsize=14)
ax.set_ylabel('Fitness', fontsize=14)
ax.set_title('Hill-Valley Method Visualization', fontsize=16)
ax.legend(fontsize=12)
ax.grid(True)

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Define the fitness landscape
x = np.linspace(-3, 3, 400)
y = x**4 - 8*x**2 + 3*x + 20

fig, ax = plt.subplots(figsize=(10, 6))

# Plot the fitness landscape
ax.plot(x, y, label='Fitness Landscape', color='blue')

# Highlight global minimum
global_min_x = x[np.argmin(y)]
global_min_y = np.min(y)
ax.scatter(global_min_x, global_min_y, color='red', s=100, zorder=5)
ax.text(global_min_x, global_min_y - 3, 'Global Minimum', fontsize=12, ha='center', color='red')

# Highlight local minima
local_minima_x = [-2, 2]
local_minima_y = [(-2)**4 - 8*(-2)**2 + 3*(-2) + 20, 2**4 - 8*2**2 + 3*2 + 20]
ax.scatter(local_minima_x, local_minima_y, color='green', s=100, zorder=5)
for xi, yi in zip(local_minima_x, local_minima_y):
    ax.text(xi, yi + 2, 'Local Minimum', fontsize=12, ha='center', color='green')

# Annotations
ax.annotate('Global optimum', xy=(global_min_x, global_min_y), xytext=(global_min_x, global_min_y - 10),
            arrowprops=dict(facecolor='red', arrowstyle='->'),
            fontsize=12, color='red', ha='center')

ax.annotate('Local optima', xy=(2, local_minima_y[1]), xytext=(2, local_minima_y[1] + 10),
            arrowprops=dict(facecolor='green', arrowstyle='->'),
            fontsize=12, color='green', ha='center')

# Labels and title
ax.set_xlabel('Search Space', fontsize=14)
ax.set_ylabel('Fitness', fontsize=14)
ax.set_title('Visualization of Local and Global Optima', fontsize=16)
ax.legend(fontsize=12)
ax.grid(True)

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Define the fitness landscape using a combination of sine and cosine functions
x = np.linspace(0, 10, 500)
y = np.sin(x) + np.cos(0.7*x)

fig, ax = plt.subplots(figsize=(10, 6))

# Plot the fitness landscape
ax.plot(x, y, color='gray')

# Define points similar to the image provided
points = {
    'x1': (2, np.sin(2) + np.cos(0.7*2)),
    'x*': (4, np.sin(4) + np.cos(0.7*4)),
    'x2': (5.5, np.sin(5.5) + np.cos(0.7*5.5)),
    'x3': (7.5, np.sin(7.5) + np.cos(0.7*7.5)),
}

# Plot the marked points with diamond shapes and labels
for label, (px, py) in points.items():
    ax.scatter(px, py, marker='D', color='black', s=80, zorder=5)
    ax.text(px, py + 0.2, f'${label}$', fontsize=14, ha='center')

# Add some random intermediate points as black dots
intermediate_x = [1, 3, 6, 8.5]
intermediate_y = [np.sin(xi) + np.cos(0.7*xi) for xi in intermediate_x]
ax.scatter(intermediate_x, intermediate_y, color='black', s=40)

# Labels and titles
ax.set_xlabel('$x$', fontsize=14)
ax.set_ylabel('$eval(x)$', fontsize=14)
ax.set_title('Visualization of fitness landscape with points $x_1$, $x_2$, $x_3$ and $x^*$', fontsize=16)

# Removing x,y ticks to match style
ax.set_xticks([])
ax.set_yticks([])

# Add grid for clarity
ax.grid(True, linestyle='--', alpha=0.5)

plt.show()
```
